# 'Menganalisis' Data Chat

````python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
````

## Membaca Data

Untungnya, LINE memiliki fasilitas untuk meng-*export* data *chat history* sebagai file `.txt`. Sayangnya, saya tidak dapat mengakses *history* pesan yang lebih lama daripada tanggal 23 November 2019.. mungkin keterbatasan dari pihak LINE? Apapun itu, biarkanlah.

Ketika melihat file *history*, kita bakal menjumpai isi yang kurang lebih sebagai berikut:

```
Wed, 01/04/2020
12:34\tkeka.vigi\tpesan satu baris
12:43\tkeka.vigi\tpesan dengan multi baris
akan tercatat seperti ini.
21:34\tkeka.vigi\t[Photo]
21:43\tkeka.vigi\t[Sticker]

Thu, 02/04/2020
23:14\tkeka.vigi removed keka.vigi from group.
```

Untuk mengubah data seperti di atas menjadi data *time series*, ada beberapa cara untuk melakukannya... salah satunya:

````python
data_by_date = {}
day = ''

with open("Chat history.txt", "r", buffering=1) as file:
    # baca file baris demi baris untuk menghemat memori :)
    for line in file.buffer:
        line = line.decode('utf-8')
        
        # cek apakah line adalah penanda hari, misal "Wed, 01/04/2020" 
        if len(line)==16 and (line[3], line[7], line[10])==(',', '/', '/'):
            day = line[:-1]
            data_by_date[day] = []
        
        # kalau bukan, berarti line adalah data chat
        elif day in data_by_date:
            
            # cek apakah line adalah pesan chat yang baru
            # misalnya "12:34\tkeka.vigi\ttesting..."
            if len(line)>5 and (line[2], line[5])==(':', '\t'):
                # split pesan menjadi ["12:34", "keka.vigi", "testing..."]
                data_by_date[day].append(line.split('\t'))
            
            # kalau tidak, line adalah kelanjutan dari chat sebelumnya
            # tambahkan line ke data sebelumnya...
            else: data_by_date[day][-1][-1] += line
````

Variabel `data_by_date` baru mengelompokkan post berdasarkan tanggal...

````python
timeline = {}

for day, posts in data_by_date.items():
    
    for hour, *action in posts:
        
        ## ingat, ada dua tipe data yang tercatat di file ini:
        ## ["12:34", "keka.vigi", "testing..."]; dan
        ## ["23:14", "keka.vigi changed group picture"]
        ##
        ## saya hanya peduli dengan data yg punya bentuk pertama, jadi
        ## jika action = ["keka.vigi changed group picture"], skip aja :)
        if len(action)==1: continue

        timeline["{} {}".format(day, hour)] = {'by':action[0], 'text':action[1][:-1]}
````

````python
timeline['Wed, 01/01/2020 00:48']
````

Dengan pandas, kita dapat membentuk *time series*

````python
data = pd.DataFrame.from_dict(timeline, orient='index')
data.index = pd.to_datetime(data.index, format="%a, %d/%m/%Y %H:%M")

data.head()
````

## Eksperimen

Hal pertama yang saya ingin ketahui, berapa banyak sih jumlah chat di grup tiap harinya? Hal ini dapat dilakukan dengan melakukan *resampling* data menjadi per hari, lalu menghitung banyak chat (cukup retoris... mungkin?)

````python
dam = data.resample('D').count()
dam.plot.line(figsize=(10, 4));
````

BTW, fungsi `len()` adalah fungsi standar di Python untuk menghitung banyak elemen dari suatu data. Sebenarnya/Dengan demikian, `data.resample('D').count()` juga dapat ditulis sebagai `data.resample('D').apply(len)` [1]. Trivia ini berguna jika kita penasaran dengan banyaknya orang yang ngechat di suatu hari. Kita tinggal meng-*apply* sebuah fungsi yang dapat menghitung banyak data yang unik dari hasil *resampling*:


[1] Dengan asumsi tidak ada data `NaN` di data sih :)

````python
dam = data.resample('D').apply(pd.Series.nunique)
dam.plot.line(figsize=(10, 4));
````

Selanjutnya, bagaimana kita melihat daftar anggota yang paling aktif di grup?

````python
# kelompokkan data berdasarkan anggota dan summary-kan
dam = data.groupby("by").describe()
````

````python
dam.head()
````

*saya gatau bagaimana mengintepretasikan `top` dan `freq`, mari buang data itu saja*
Selanjutnya, kita tinggal mengurutkan data berdasarkan banyak teks, lalu berdasarkan banyak teks yg unik (dalam kasus ada orang yang seri)

````python
dum = dam['text'].sort_values(by=["count", "unique"], ascending=False)

dum = dum[['count', 'unique']]
dum.head(n=10)
````
