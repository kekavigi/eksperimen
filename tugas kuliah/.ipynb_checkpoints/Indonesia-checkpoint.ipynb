```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
from scipy.optimize import minimize

from matplotlib import rc
rc('text', usetex=True)

for zzz in range(10, 25):
    print(zzz)
    # https://docs.google.com/spreadsheets/d/1ma1T9hWbec1pXlwZ89WakRk-OfVUQZsOCFl4FwZxzVw/htmlview?usp=sharing#
    infected  = [2, 2, 2, 2, 4, 4, 6, 19, 27, 30, 27, 60, 83, 104, 121, 156, 197, 269, 321, 392, 437, 500, 601, 701][:zzz]#
    recovered = [0, 0, 0, 0, 0, 0, 0,  0,  0,  4,  7,  9, 13,  13,  13,  16,  30,  40,  48,  58,  77,  79,  85,  89][:zzz]#
    day = len(infected)
    kasus = [infected[t]+recovered[t] for t in range(day)]

    def plotSIR(N, beta, gamma):   
        def SIR(t,y):
            S, I, R = y
            return([-beta*S*I, beta*S*I-gamma*I, gamma*I])
        sol = solve_ivp(SIR, [0, 60], [N, 2, 0], t_eval=np.arange(0, 60+.2, 0.2))

        plt.figure(figsize=(16, 10))
        plt.axis([0, 60, 0, 2000])
        plt.plot(sol.t, sol.y[1])
        plt.plot(sol.t, sol.y[2])
        plt.plot(sol.t, sol.y[1] + sol.y[2])
        plt.plot(np.arange(0, day), infected , "k*:")
        plt.plot(np.arange(0, day), recovered, "k*:")
        plt.plot(np.arange(0, day), kasus    , "k*:")

        now = infected[-1]
        plt.annotate(now, (day-1.4, now+20))
        now = recovered[-1]
        plt.annotate(now, (day-1.4, now+20))
        now = kasus[-1]
        plt.annotate(now, (day-1.4, now+20))

        for i in [1, 2]:
            predict = sol.y[i][::5][day]
            plt.plot(day, predict, 'ko')
            plt.annotate(str(int(round(predict))), (day-0.3, predict+25))

            if i==2: pass
            maksi = max(sol.y[i][::5])
            when = sol.t[list(sol.y[i]).index(maksi)]
            plt.plot(when, maksi, 'ko')
            plt.annotate(str(int(round(maksi))), (when-0.3, maksi+25))

        predict = (sol.y[1]+sol.y[2])[::5][day]
        plt.plot(day, predict, 'ko')
        plt.annotate(str(int(round(predict))), (day-0.3, predict+25))

        plt.minorticks_on()
        plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.3)
        plt.grid(b=True, which='major', color='#666666', linestyle='-')

        plt.legend(["Case", "Infected", 'Recovered', "$Data_C$", "$Data_I$", "$Data_R$"])
        plt.title('Day {}'.format(day))
        plt.savefig('line_plot_{}.png'.format(day))  

    def SSE(p):
        N, beta, gamma = p
        def SIR(t,y):
            S, I, R = y
            return([-beta*S*I, beta*S*I-gamma*I, gamma*I])

        sol = solve_ivp(SIR, [0, day], [N, 2, 0], t_eval=np.arange(0, day+.2, 0.2))
        sseinfected  = sum((sol.y[1][:-1:5]-infected )**2)
        sserecovered = sum((sol.y[2][:-1:5]-recovered)**2)
        return(sseinfected+sserecovered)


    msol = minimize(SSE, [10000, 0.001, 1], method='Nelder-Mead')
    N, beta, gamma = msol.x
    plotSIR(N, beta, gamma)
```
