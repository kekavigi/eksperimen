```python
import pandas as pd
import numpy as np
```

```python
digi = pd.read_csv('digilib.tsv', sep='\t')
digi = digi.replace(np.nan, '', regex=True)

digi['dc:nim'] = digi['dc:nim'].astype(str)
```

```python
bynim = digi[digi['dc:nim'].str[:3]=='101']
bysub = digi[digi['subject']=='FMIPA - Matematika']

union = pd.concat([bynim, bysub], join='outer')
union = union.reset_index()
union = union.drop(columns=['Unnamed: 0', 'index'])
```

```python
union.to_csv('fmipa-ma.tsv', sep='\t')
```
